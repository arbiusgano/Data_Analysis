## **Task 1 : Cleaning The Data**

```python
dataset.dtypes
```

```python
dataset["Date"]=dataset["Date"].astype("datetime64")
dataset["Region"]=dataset["Region"].astype("category")
dataset["Type"]=dataset["Type"].astype("str")
dataset["Units"]=dataset["Units"].astype("int")
```

```python
## The dataset cannot read because the dataset have NaN
dataset.isnull().sum()
```

```python
dataset.isna().sum()
```

```python
## Fill the Missing data 
dataset_fillna=dataset.copy()
dataset_fillna["Units"]=dataset_fillna["Units"].fillna(dataset_fillna["Units"].median())
```

```python
dataset_fillna["Date"]=dataset_fillna["Date"].astype("datetime64")
dataset_fillna["Region"]=dataset_fillna["Region"].astype("category")
dataset_fillna["Type"]=dataset_fillna["Type"].astype("str")
dataset_fillna["Units"]=dataset_fillna["Units"].astype("int")
```
