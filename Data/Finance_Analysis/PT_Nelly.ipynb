# ***CheckPoint***

```python
df=cleaning_data.copy()
```

```python
df.head()
```

```python
# Gross Profit Margin (GPM) = (Laba Kotor : Total Pendapatan) x 100%
df["GPM %"]=df["PENJUALAN BERSIH"]/df["PENJUALAN "]*100
```

```python
# Profit Margin Ratio (PMR) or Net Profit Margin (NPM) = Laba Bersih Setelah Pajak : Penjualan
df["NPM %"]=df["LABA SETELAH PAJAK"]/df["PENJUALAN "]
```

```python
# Return on Assets Ratio (ROA) =  Laba Bersih : Total Aset
df["ROA %"]=df["LABA SEBELUM PAJAK"]/df["TOTAL ASET"]
```

```python
# Return on Equity Ratio (ROE) = Laba Bersih Setelah Pajak : Ekuitas Pemegang Saham
df["ROE %"]=df["LABA SEBELUM PAJAK"]/df["EKUITAS"]
```

```python
# Return on Sales Ratio (ROS) = (Laba sebelum pajak dan bunga : penjualan) x 100%
df["ROS %"]=df["LABA SEBELUM PAJAK"]/df["PENJUALAN "]*100
```
