```python
import pandas as pd
```

```python
df = pd.read_csv ('./data_cleaning_challenge.csv')
```

```python
df
```

```python
drop_extra = df.drop(columns=['Unnamed: 9','Unnamed: 10'])
```

```python
drop_extra
```

```python
drop_extra
```

```python
drop_nan = drop_extra[drop_extra["Row Type"].notna()]
```

```python
df = drop_nan
```

```python
df
```

```python
collumn_value = []
counter = 0

for i in df ['Row Type']:
    if "first name" in i:
        counter += 1
    collumn_value.append(counter)
```

```python
collumn_value
```

```python
df['iteration']=collumn_value
```

```python
df
```

```python
drop_extra_collumn = df[df['Row Type'] !="Row Type"]
```

```python
drop_extra_collumn.tail()
```

```python
Name_Df = drop_extra_collumn[drop_extra_collumn['Row Type'].str.contains('first name')]
```

```python
Name_Df.head()
```

```python
Name_Df
```

```python
Name_Df.drop(columns=['Speed1','Speed2','Electricity','Effort','Weight','Torque'],inplace=True)
```

```python
Name_Df
```

```python
Name_Df.drop(columns=['Torque'],inplace=True)
```

```python
Name_Df
```

```python
Name_Df.rename(columns={"Row Type":"First Name","Iter Number":"Last Name","Power1":"Date"},inplace=True)
```

```python
Name_Df["First Name"]=Name_Df["First Name"].str[12:]
Name_Df["Last Name"]=Name_Df["Last Name"].str[11:]
Name_Df["Date"]=Name_Df["Date"].str[6:]
```

```python
Name_Df
```

```python
No_Name_Df = drop_extra_collumn[~drop_extra_collumn['Row Type'].str.contains('first name')]
No_Name_Df
```

```python
pd.merge(left=Name_Df, right=No_Name_Df, how="inner", on="iteration")
```
