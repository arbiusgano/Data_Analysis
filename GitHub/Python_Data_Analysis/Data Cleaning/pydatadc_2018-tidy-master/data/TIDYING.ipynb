## Permasalahan variabel ada di collomn dan row 

```python
weather = pd.read_csv("./weather.csv")
```

```python
weather
```

```python
weather_melt = pd.melt(weather,id_vars=["id","year","month","element"],
                               var_name="day",
                               value_name="temp"
)
```

```python
weather_melt.head()
```

# Pivot Table
## Kita mau nambah kolom baru 
## yang isinya tmax & tmin dari kolom element 

```python
weather_tidy = pd.pivot_table(weather_melt,
                              index=["id","year","month","day"],  
                              columns="element",
                              values="temp")
```

```python
weather_tidy.head()
```

## Kalua kita mau ubah biar terlihat seperti yg weather_melt

```python
weather_tidy.reset_index()
```

```python
billboard = pd.read_csv("./billboard.csv")
```

```python
billboard.head()
```

```python
billboard_melt=pd.melt (
    billboard,
    id_vars=['year','artist','track','time','date.entered'],
    var_name='week',
    value_name='rating'
)
```

```python
billboard_melt.head()
```

## Buat tabel baru dari tabel yang ada tapi kolomnya hanya sebagian yg diambil
## lalu kita hapus data yg duplikat karena datanya diambil dari data yg sdh di melt

```python
billboard_melt.head()
```

```python
billboard_song = billboard_melt[["year","artist","track","time","date.entered"]]
```

```python
billboard_song
```

## Disini kita mau drop data yang duplikat 

```python
billboard_song.shape
```

```python
billboard_song_drop_duplicate = billboard_song.drop_duplicates()
```

## Sekarang menjadi 317 row dari awalnya 24092 row

```python
billboard_song_drop_duplicate
```

## Bikin kolom baru ID biar bisa di join sama kolom rating 


```python
billboard_song_drop_duplicate ['ID'] = range(len(billboard_song_drop_duplicate))  
```

```python
billboard_song_drop_duplicate.head()
```

## Save data ke bentuk csv yg sudah di cleaning & tidying
## pakai index=False biar indexnya tidak ikut tesave

```python
billboard_song_drop_duplicate.to_csv('billbord_song.csv', index=False)  
```
