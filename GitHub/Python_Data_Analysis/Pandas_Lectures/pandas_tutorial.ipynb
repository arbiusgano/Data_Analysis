# splitting Collums 
## (membagi kolom misal di kolom date_added [march 1,2020] kta mau pisah data ygdi kolom berdasarkan komanya jadi nanti akan ada  kolom baru yang hanya terdiri dari tanggal bulan, dan tahunnya saja)

```python
netflix [["Bulan & Tanggal Ditambahkan","Tahun ditambahkan"]] = netflix["date_added"].str.split(", ",expand=True)
```

```python
netflix
```

```python
netflix = netflix.drop (columns="date_added")
```

# Rename Collum (ganti nama kolom)
## Rename lebih dari 1 kolom tinngal dikasih koma 

```python
netflix.rename( columns= {"Judul": "judul","Tahun ditambahkan": "tahun_ditambahkan"}, inplace=True)
netflix.rename( columns= {"listed_in": "genre"}, inplace=True)
```

```python
netflix
```

# Mengisi data yang Null di Python (NaN)

```python

netflix["country"].fillna(value="no_country",inplace=True)
```

```python
netflix
```

# menghitung karakter yang ada pada tiap baris 
## (disini contoh hitung genre yang ada di genre)

```python
netflix["jumlah_genre"] = netflix["genre"].map(lambda x: len(x.split(",")))

netflix["jumlah_pemeran"] = netflix["pemeran"].map(lambda x: len(x.split(",")))
```
