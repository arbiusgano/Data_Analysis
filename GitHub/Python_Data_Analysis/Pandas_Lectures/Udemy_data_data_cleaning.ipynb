### Creat new colloum Day of the week

```python
## So on the DF we has know the tanggal but we dont know what day its on week 
## So the goals is to exctact the day of the week
## And here we will define fungtions date_to_weekday
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
## And next we just need to use the fungtions thas we has def(define)
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```
