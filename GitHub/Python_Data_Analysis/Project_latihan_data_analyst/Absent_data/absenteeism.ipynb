```python
## We need to create the variabel so we can put in df_drop_ID 
## Use axis 1 count it horizontaly
reason_type_1 = reason_column.loc [:,1:14].max(axis=1)
reason_type_2 = reason_column.loc [:,15:17].max(axis=1)
reason_type_3 = reason_column.loc [:,18:21].max(axis=1)
reason_type_4 = reason_column.loc [:,22:28].max(axis=1)
```

## Next Task Working With Date Colloum

```python
df_coloumn=df_clean.copy()
```

```python
df_coloumn_only = df_coloumn["Date"]
df_coloumn_only.info
```

```python
## Main Problem is date_colloum use / for seperating day,month,year | So we need change thats so can being analyst 
## Or the datatype is not float or integer 
df_clean["Date"]=pd.to_datetime (df_clean["Date"])
```

## Extract The Month Value : 

```python
## Append is excracting the data so in this case we want to exctract month value only but from 700 rows 
## So we will use for loop and storing the value to a list, and thas is why we create variabel list_month = []
list_month = []

for i in range(700):
    list_month.append(df_clean["Date"][i].month)
```

```python
df_clean["Month Value"]=list_month
df_clean
```

### Creat new colloum Day of the week

```python
## So on the DF we has know the tanggal but we dont know what day its on week 
## So the goals is to exctact the day of the week
## And here we will define fungtions date_to_weekday
def date_to_weekday (date_value):
    return date_value.weekday()
```

```python
## And next we just need to use the fungtions thas we has def(define)
df_clean["Day of Weekday"]=df_clean["Date"].apply(date_to_weekday)
```
