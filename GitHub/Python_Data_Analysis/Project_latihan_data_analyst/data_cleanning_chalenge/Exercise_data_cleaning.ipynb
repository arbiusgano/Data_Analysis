```python
import pandas as pd
```

## Data Cleaning Exercise
##### Datasetnya terdiri dari banyak tabel dengan kolom yang sama
##### Buat kolom baru yang isinya First Name, Last Name, Date
##### Barisnya untuk First Name, Last Name, Date kita ambil dari baris 0 Jadi nanti kita buat Iteration
##### Dibagungkan dengan Tabel Row Type sampai Torque   
##### Format kolom akhir (First Name,Last Name,Date,Row Type,Iter Number,Power1,Speed1,Speed2,Effort,Weight,Torque)


```python
df_raw=pd.read_csv("./data_cleaning_challenge.csv",delimiter=",")
df_raw.head()
```

```python
### Drop extra kolom
### Baris 1 dari setiap tabel
drop_extra=df_raw.drop(columns=["Unnamed: 9","Unnamed: 10"])
df_1=drop_extra[drop_extra["Row Type"].notna()]
df_1.head()
```

```python
### Kita buat Iteration
kolom_value=[]
counter=0

for i in df_1["Row Type"]:
    if "first name" in i:
        counter+=1
    kolom_value.append(counter)
```

```python
### Drop baris 2 pada setiap tabel
df_1=df_1[df_1["Row Type"]!="Row Type"]

df_1.head()
```

```python
### Kita masukan tabel Iteration ke dataset
df_1["Iteration"]=kolom_value
```

```python
### error = ValueError: Length of values (58397) does not match length of index (52403)
### Sorry Ada error karena kita urutan prosesnya tertukar harusnya tambah kolom dlu baru drop Rows 2 td
### Jadi kita buat ulang lg Iteration karena variabelnya sama jadi nga bisa ulang hehe
kolom_value=[]
iteration=0

for i in df_1["Row Type"]:
    if "first name" in i:
        iteration +=1
    kolom_value.append(iteration)
```

```python
### Kita masukan tabel Iteration ke dataset
df_1["Iteration"]=kolom_value
df_1
```

```python
### Sekarang kita buat kolom baru First Name,Last Name,Date
### Dan Isi kolom baru dari rows 0 pada tiap tabel 
name_data=df_1[df_1["Row Type"].str.contains("first name")]

name_data.head()
```

```python
### Kita drop kolom Speed1 - Torque 
### Karena nanti kita akan gabung degan df1 tadi 
### Dan kita Rename kolom Row Type, Iter Number, Power menjadi First Name, Last Name, Date
name_data.columns
```

```python
name_data.drop(columns=['Speed1', 'Speed2', 'Electricity',
       'Effort', 'Weight', 'Torque'],inplace=True)
name_data.rename(columns={'Row Type':'First Name', 'Iter Number':'Last Name', 'Power1':'Date'},inplace=True)

name_data
```

```python
### Oke sekarang kita mau hapus isi tabel tiap kolom dan hanya menyiskan Person, Human, dan end of time
name_data["First Name"]=name_data["First Name"].str[12:] 
### Jadi kita mau hapus string 1-12 pada semua row pada kolom "First Name" kenapa 12 karena python hitung mulai dari 0
name_data["Last Name"]=name_data["Last Name"].str[11:] 
name_data["Date"]=name_data["Date"].str[6:] 

name_data.head()
```

```python
### Okee good2 Clean sekarang tinggal di gabungkan dengan tabel df1 td
noname_data=df_1[~df_1["Row Type"].str.contains("first name")]

noname_data.head()
```

```python
final_df = pd.merge(left=name_data,right=noname_data,how="inner",on="Iteration")
final_df.head()
```

```python
final_df.drop(columns=["Iteration"],inplace=True)
final_df
```

```python
### Tinggal di save ke CSV atau ke SQL databes
### Indexnya setting false biar tidak ikut ke save
final_df.to_csv("Data_Cleaning_Exercise.csv",index=False)
```
